Assignment

Generate a rolling window for adjacent close, opening,low and high prices of the stock for 5 companies,

```python
from datetime import datetime
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
import yfinance as yfin
yfin.pdr_override()

def Get_Stock_Data(Ticker,start,end):
    data = pdr.get_data_yahoo(Ticker, start, end)
    data.insert(0,"Ticker",Ticker)
    return data

start = datetime(2023,1,1)
end = datetime.today()

ibm_data = Get_Stock_Data('IBM',start,end)
apple_data = Get_Stock_Data('AAPL',start,end)
google_data = Get_Stock_Data('GOOGL',start,end)
amazon_data = Get_Stock_Data('AMZN',start,end)
netflix_data = Get_Stock_Data('NFLX',start,end)

rolling_window_5_ibm = ibm_data['Adj Close'].rolling(window=5).mean()
rolling_window_5_apple = apple_data['Adj Close'].rolling(window=5).mean()
rolling_window_5_google = google_data['Adj Close'].rolling(window=5).mean()
rolling_window_5_amazon = amazon_data['Adj Close'].rolling(window=5).mean()
rolling_window_5_netflix = netflix_data['Adj Close'].rolling(window=5).mean()


plt.title("IBM")
plt.plot(ibm_data['Adj Close'], label = "adj close", linewidth = 4)
plt.plot(rolling_window_5_ibm, label='rolling window 5')
plt.legend()
plt.show()
plt.title("Apple")
plt.plot(apple_data['Adj Close'], label = "adj close", linewidth = 4)
plt.plot(rolling_window_5_apple, label='rolling window 5')
plt.show()
plt.title("Google")
plt.plot(google_data['Adj Close'], label = "adj close", linewidth = 4)
plt.plot(rolling_window_5_google, label='rolling window 5')
plt.show()
plt.title("AMZN")
plt.plot(amazon_data['Adj Close'], label = "adj close", linewidth = 4)
plt.plot(rolling_window_5_amazon, label='rolling window 5')
plt.show()
plt.title("NETFLIX")
plt.plot(netflix_data['Adj Close'], label = "adj close", linewidth = 4)
plt.plot(rolling_window_5_netflix, label='rolling window 5')
plt.show()
```
