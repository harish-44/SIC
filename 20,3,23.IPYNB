```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy

dataframe = pd.read_csv("/workspaces/SIC/allCoountries.csv")
print(dataframe.head(10))
```

```python
dataframe.drop(['Code'],axis = 1,inplace = True) 
dataframe.head(10)
```

```python
#Financial Data Analysis 
from datetime import datetime
import pandas as pd
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
import yfinance as yfin #Yahoo Finance
yfin.pdr_override() 

start = datetime(2020,1,1)
end = datetime.today()

def Get_Stock_Data(Ticker,start,end): #Ticker is the stock symbol
    data = pdr.get_data_yahoo(Ticker, start, end) #Get the data
    data.insert(0,"Ticker",Ticker) #Insert the Ticker as the first column
    return data

ticker = ["IBM","CTS","TCS","BIS","SPY"]

dataframe1 = pd.DataFrame()
dataframe1 = Get_Stock_Data(ticker[0],start,end)
dataframe2 = pd.DataFrame()
dataframe2 = Get_Stock_Data(ticker[1],start,end)
dataframe3 = pd.DataFrame()
dataframe3 = Get_Stock_Data(ticker[2],start,end)
dataframe4 = pd.DataFrame()
dataframe4 = Get_Stock_Data(ticker[3],start,end)
dataframe5 = pd.DataFrame()
dataframe5 = Get_Stock_Data(ticker[4],start,end)
dataframe1
dataframe2
dataframe3
dataframe4
dataframe5
```

```python
dataframe1= dataframe1.pivot(index= None,columns = 'Ticker',values = 'Adj Close')
dataframe1
```

```python
dataframe2= dataframe2.pivot(index= None,columns = 'Ticker',values = 'Adj Close')
dataframe2
```

```python
dataframe3= dataframe3.pivot(index= None,columns = 'Ticker',values = 'Adj Close')
dataframe3
```

```python
dataframe4 = dataframe4.pivot(index= None,columns = 'Ticker',values = 'Adj Close')
dataframe4
```

```python
dataframe5 = dataframe5.pivot(index= None,columns = 'Ticker',values = 'Adj Close')
dataframe5
```

```python
dataframe1.plot()
```

```python
dataframe = pd.DataFrame()
dataframe = pd.concat([dataframe1,dataframe2,dataframe3,dataframe4,dataframe5],axis = 1,join ='outer')
dataframe
```
